```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv file and add column as X and Y
df = pd.read_csv("../dataset/data/experimental/bruno_data.csv", header=None, names=['X', 'Y'])
# range of X and Y
print(df['X'].min(), df['X'].max())
print(df['Y'].min(), df['Y'].max())
```

```python
# plot scatter plot X and Y but I want Y to be reversed (range from maximum to minimum). I want marker with facecolor='none' and edgecolor='black'
plt.figure(figsize=(4, 4))
plt.scatter(df['X'], df['Y'], facecolor='none', edgecolor='black')
plt.gca().invert_yaxis()  # this is the line that inverts the y-axis
x_ticks = list(range(2, 21, 4))  # [2, 4, 6, ..., 20]
plt.xticks(x_ticks,fontsize=12)
plt.xlim(0, 22)

y_ticks = list(range(0, 2001, 200))  # [0, 200, 400, ..., 2000]
y_labels = [1800, 1400, 1000, 600, 200]
y_ticklabels = [str(y) if y in y_labels else '' for y in y_ticks]
plt.yticks(y_ticks, y_ticklabels, fontsize=12)
plt.xlabel('PSP onset latency (ms)', fontsize=12)
plt.ylabel('Microdrive depth ($\mu$m)', fontsize=12)
# remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
# remove white background
plt.gca().set_facecolor('none')
plt.savefig('saved_figs/fig1_latency.svg', dpi=300, bbox_inches='tight')
plt.show()

```
